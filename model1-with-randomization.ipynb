```python
import pandas as pd

raw = pd.read_csv('datasets/fifa_22_preprocessed.csv') #read in data
```

#### drop colums I'm not using

```python
data = raw.drop(['Unnamed: 0', 'Name', 'Club'], axis=1) # not useful as predictors

data = data.drop(['PaceTotal', 'ShootingTotal', 'PassingTotal',
                'DribblingTotal', 'DefendingTotal', 'PhysicalityTotal',
                'BaseStats', 'TotalStats', 'WageEUR'],
               axis=1)
        # removing totals because I'm randomizing sub-stats that comprise totals
        # in order to simulate moe on scout evaluation of players
        
        # also removing Wages as this info is commonly not public knowledge
```

#### create random column

```python
# features to randomize
rand_features = ['Crossing', 'Finishing', 'HeadingAccuracy', 'ShortPassing', 'Volleys',
       'Dribbling', 'Curve', 'FKAccuracy', 'LongPassing', 'BallControl',
       'Acceleration', 'SprintSpeed', 'Agility', 'Reactions', 'Balance',
       'ShotPower', 'Jumping', 'Stamina', 'Strength', 'LongShots',
       'Aggression', 'Interceptions', 'Positioning', 'Vision', 'Penalties',
       'Composure', 'Marking', 'StandingTackle', 'SlidingTackle', 'GKDiving',
       'GKHandling', 'GKKicking', 'GKPositioning', 'GKReflexes', 'STRating',
       'LWRating', 'LFRating', 'CFRating', 'RFRating', 'RWRating', 'CAMRating',
       'LMRating', 'CMRating', 'RMRating', 'LWBRating', 'CDMRating',
       'RWBRating', 'LBRating', 'CBRating', 'RBRating', 'GKRating']
```

```python
# values before randomization
data[rand_features].head()
```

```python
from numpy import random as r

for f in rand_features:
        data[f] = r.randint(data[f] - 10, data[f] + 10)
```

```python
# values after randomization
data[rand_features].head()
```

```python
y = data['ValueEUR'] # set target variable
X = data.drop('ValueEUR', axis=1) # set features
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.8, test_size=0.2)
```

```python
from sklearn.pipeline import Pipeline
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error

model = RandomForestRegressor(n_estimators=100, random_state=0)

my_pipeline = Pipeline(steps=[('model', model)])

my_pipeline.fit(X_train, y_train)

preds = my_pipeline.predict(X_test)

score = mean_absolute_error(y_test, preds)
print('MAE:', score)
```
